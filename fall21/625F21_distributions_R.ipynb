# Visualizing Distributions Examples in R

## Setup

```
library(tidyverse)
library(ggplot2)
library(scales)        # allows us to format axes labels with commas
options(scipen=999)    # prevent using scientific notation
theme_set(
  theme_bw()   # selects a black and white theme
  + theme(text = element_text(size = 20))  # sets larger font size
  + theme(plot.margin = margin(0.5, 1, 0, 0, "cm"))
#  + theme (plot.background = element_rect(fill = "grey90", colour = "black", size = 1)))
)
```

```
# function to set plot sizes in notebook
# ex: fig (10,4)
# https://www.kaggle.com/getting-started/105201
fig <- function(width, heigth){
     options(repr.plot.width = width, repr.plot.height = heigth)
}
```

Use dataset of diamond prices built into ggplot2  
https://ggplot2.tidyverse.org/reference/diamonds.html

```
data("diamonds")
```

```
head(diamonds)
```

## Histogram

For the histogram, we show the distribution of diamond prices.  

To create the histogram, we use the `geom_histogram()` function.  `binwidth` sets the size of each histogram bin (this one is set to 500, representing $500).

```
fig(10,6)
ggplot(diamonds, aes(price)) + 
  geom_histogram(binwidth=500) +
  scale_x_continuous(label=comma) +  # use commas in x axis labels
  scale_y_continuous(label=comma) +  # use commas in y axis labels
  labs(x = "price ($)", caption="source: diamonds")
```

Changing the binwidth to $1000

```
fig(10,6)
ggplot(diamonds, aes(price)) + 
  geom_histogram(binwidth=1000) +
  scale_x_continuous(label=comma) +
  scale_y_continuous(label=comma) +  # use commas in y axis labels
  labs(x = "price ($)", caption="source: diamonds")
```

Changing the binwidth to $100

```
fig(10,6)
ggplot(diamonds, aes(price)) + 
  geom_histogram(binwidth=100) +
  scale_x_continuous(label=comma) +
  scale_y_continuous(label=comma) +  # use commas in y axis labels
  labs(x = "price ($)", caption="source: diamonds")
```

## Boxplot

We're again looking at the price of diamonds.  This time, we'll use boxplots (`geom_boxplot()`).

```
fig(4,6)
ggplot(diamonds, aes(y=price)) + 
  scale_y_continuous(label=comma) +
  geom_boxplot() + 
  labs(y = "price ($)", caption="source: diamonds")
```

Now, let's look at these by cut and have a separate box plot for each cut type.

```
fig(12,6)
ggplot(diamonds, aes(x=cut, y=price)) + 
  scale_y_continuous(label=comma) +
  geom_boxplot() + 
  labs(y = "price ($)", caption="source: diamonds")
```

## Empirical CDF

For the empirical CDF (CDF), we just have to take the histogram code and replace `geom_histogram()` with `stat_ecdf()`.

```
fig(10,6)
ggplot(diamonds, aes(price)) + 
  stat_ecdf() +
  scale_x_continuous(label=comma) +
  labs(x = "price ($)", caption="source: diamonds")
```
